Usaremos la librería NertworkX

```python
# Librerías
# ======================================================================================
import pandas as pd
import warnings
import matplotlib.pyplot as plt

import networkx as nx

warnings.filterwarnings("ignore")
```

Para crear un grafo vacio usamos la función nx.Graph()

```python
# Creación de una instancia de tipo "Grafo"
# ======================================================================================
G = nx.Graph()
print(G)
```

¿Cómo comprobamos si es un grafo dirigido o no?

```python
if G.is_directed():
    print("El grafo es dirigido")
else:
    print("El grafo no es dirigido")
```

Para añadir un nodo usamos *add_node*.
Crea un nodo que se llame "Pla d'en Boet"

```python
fig, ax = plt.subplots(figsize=(2,2))
G.add_node("Pla d'en Boet")
nx.draw(G, with_labels=True, ax=ax)
print(G)
```

Para añadir varios nodos a la vez usamos *add_nodes_from*


```python
# Añadir multiples nodos
# ======================================================================================
nodos = ["Eixample","Cerdanyola", "La Llàntia", "Cirera", "Rocafonda","Vista Alegre", "El palau Escorxador","Peramàs","Els Molins", "Centre"]
G.add_nodes_from(nodos)

fig, ax = plt.subplots(figsize=(5, 5))
nx.draw(G, with_labels=True, ax=ax)
ax.set_xlim([1.2*x for x in ax.get_xlim()])
ax.set_ylim([1.2*y for y in ax.get_ylim()])
print(G)
```

Para añadir un eje usamos *add_edge*.
Conecta dos barrios



```python
# Añadir un único eje
# ======================================================================================
G.add_edge("Pla d'en Boet", "Cerdanyola")

fig, ax = plt.subplots(figsize=(5, 5))
nx.draw(G, with_labels=True, ax=ax)
ax.set_xlim([1.2*x for x in ax.get_xlim()])
ax.set_ylim([1.2*y for y in ax.get_ylim()])
print(G)
```

Para añadir un eje usamos *add_edges_from*.
Conecta todos barrios


```python
# Añadir múltiples ejes
# ======================================================================================
lista_edges = [("Pla d'en Boet","Centre"),("Pla d'en Boet","Peramàs"),("Cerdanyola","Rocafonda"),("Centre","Eixample"),("Centre","El palau Escorxador"),("Vista Alegre","Eixample"),("Rocafonda","La Llàntia"),("Els Molins","El palau Escorxador"),("Vista Alegre","Cirera"),("Vista Alegre","Els Molins")]
G.add_edges_from(lista_edges)
fig, ax = plt.subplots(figsize=(10, 10))
nx.draw(G, with_labels=True, ax=ax)
ax.set_xlim([1.2*x for x in ax.get_xlim()])
ax.set_ylim([1.2*y for y in ax.get_ylim()])
print(G)
```

Contar número de arestas y número de nodos

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

Si se añade una conexión cuyos nodos no existen, se crean automáticamente.
Crea una arista de Barceloneta a Pla d'en Boet.


```python
G.add_edge("Barceloneta", "Pla d'en Boet")
fig, ax = plt.subplots(figsize=(10, 10))
nx.draw(G, with_labels=True, ax=ax)
ax.set_xlim([1.2*x for x in ax.get_xlim()])
ax.set_ylim([1.2*y for y in ax.get_ylim()])
print(G)
```

Si quiero eliminar una aresta

```python
G.remove_edge("Barceloneta","Pla d'en Boet")

fig, ax = plt.subplots(figsize=(10, 10))
nx.draw(G, with_labels=True, ax=ax)
```

Si quiero eliminar un nodo

```python
G.remove_node("Barceloneta")
fig, ax = plt.subplots(figsize=(10, 10))
nx.draw(G, with_labels=True, ax=ax)
```

Para obtener la matriz de adyacencia

```python
adj = nx.adjacency_matrix(G)
print(adj)
```

Lo quiero en formato matrix

```python
# Se convierte la matriz de formato sparse a dense para poder imprimirla
adjM = adj.todense()
print(adjM)
```

Si quiero la información del grafo de los nodos y aristas

```python
G.nodes()
```

```python
G.edges()
```

Calcular el camino más corto entre Pla d'en Boet i Vista alegre

```python
nx.shortest_path(G,"Els Molins","Rocafonda")
```

```python
G.add_node("Poble Nou")
G.add_edge("Poble Nou","Cirera",weight=2)
G.add_edge("Poble Nou","Vista Alegre",weight=23)

fig,ax = plt.subplots(figsize=(10,10))
nx.draw(G,with_labels = True,ax=ax)
nx.shortest_path(G,"Vista Alegre","Barcelona")
```
